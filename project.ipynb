This project aims to analyze customer churn data, preprocess it, build several machine learning models to predict churn, and evaluate their performance.


Customer churn (or customer attrition) is the percentage of customers who stop subscribing to or purchasing from a business over a specific period. It acts as a "leaky bucket" metric, directly reducing revenue and indicating customer dissatisfaction or competitive losses.

This section prepares the data for machine learning models by selecting relevant features and transforming them into a suitable format.

```python
#feature engineering starts here
dta.columns
```

Separates the features ('Age', 'Tenure', 'MonthlyCharges') into `x` and the target variable ('Churn') into `y`.

```python
y = dta["Churn"]
x = dta[["Age",'Tenure',"MonthlyCharges"]]
```

Converts the 'Gender' categorical column into a numerical format (1 for 'Female', 0 for 'Male'), which is required for most machine learning algorithms.

```python
x["Gender"] = dta["Gender"].apply(lambda val: 1 if val=="Female" else 0)
```

Transforms the 'Churn' target variable from categorical ('Yes', 'No') to numerical (1 for 'Yes', 0 for 'No').

```python
y = y.apply(lambda val: 1 if val == "Yes" else 0)
```

This section focuses on splitting the data, scaling features, training various machine learning models, and evaluating their performance.

Splits the dataset into training and testing sets, with 80% of the data used for training and 20% for testing. This helps in evaluating the model's performance on unseen data.

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train , y_test = train_test_split(x,y, test_size=0.2)
```

Initializes the `StandardScaler` to normalize the features. Scaling is crucial for algorithms sensitive to feature magnitudes, like KNN and SVM.

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

Applies standardization to the training data. `fit_transform` learns the scaling parameters from the training data and then applies them.

```python
x_train = scaler.fit_transform(x_train)
```

Saves the trained `StandardScaler` to a file named 'scaler.pkl'. This allows the same scaling transformation to be applied to new, unseen data.

```python
import joblib
joblib.dump(scaler,"scaler.pkl")
```

Applies standardization to the test data using the scaling parameters learned from the training data. It's important to only `transform` the test data, not `fit_transform`.

```python
x_test = scaler.fit_transform(x_test)
```

Defines a helper function `model_perfomance` to calculate and print the accuracy score of a model given its predictions and the true labels.

```python
from sklearn.metrics import accuracy_score
def model_perfomance(predictions):
  print("Acuuracy score on the model is {}".format(accuracy_score(y_test, predictions)))
```

Trains a Logistic Regression model, a linear model for binary classification, and evaluates its performance.

Trains a K-Nearest Neighbors model, a non-parametric classification algorithm, and optimizes its hyperparameters using GridSearchCV.

```python
from sklearn.neighbors import KNeighborsClassifier
```
